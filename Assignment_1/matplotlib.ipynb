```python
import numpy as np
x=np.arange(0, 100)
y=x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
ax.plot(x, y, 'b')
```

```python
figure = plt.figure()
ax1 = figure.add_axes([0, 0, 1, 1])
ax2 = figure.add_axes([0.2, 0.5, 0.2, 0.2])
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax1.plot(x, y, 'r')
ax2.plot(x, y, 'r')

plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.set_xlabel("X")
ax2.set_ylabel('Y')
ax2.set_title('zoom')

ax1.plot(x, z, 'b')
ax2.plot(x, y, 'b')

ax2.set_xlim([20, 22])
ax2.set_ylim([30, 50])

plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, linestyle='dashed', linewidth=6, color='blue')
axes[0].set_xlim([0 ,100])

axes[1].plot(x, z, linestyle='solid', linewidth=6, color='red')
axes[1].set_xlim([0, 100])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,2))

axes[0].plot(x, y, ls='solid', lw=4, color='blue')
axes[0].set_xlim([0 ,100])

axes[1].plot(x, z, ls='dashed', lw=4, color='red')
axes[1].set_xlim([0, 100])
```
