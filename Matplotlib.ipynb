```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 100, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

x = np.array([0, 1, 2, 3])
y = np.array([0 ,1, 2, 3])

axes1.plot(x, y, 'r')
axes1.set_xlabel("X_label_axes1")
axes1.set_ylabel("Y_label_axes1")
axes1.set_title("Axes label 1")

axes2.plot(x, y, 'cyan')
axes2.set_xlabel("X_label_axes2")
axes2.set_ylabel("Y_label_axes2")
axes2.set_title("Axes label 2")

plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x**2)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("A single plot")
fig.suptitle("TOKYO")

plt.show()
```

```python
fig, axs = plt.subplots(2)
# print(axs)
fig.suptitle("Vertically stacked cells")
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x**2)

axs[0].plot(x, y, 'r')
axs[0].set_title("TOKYO")
axs[1].plot(x, -y, 'p--')
axs[1].set_title("BERLIN")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle("horizontally stacked subplots")

x = np.linspace(0, 2 * np.pi, 10000000)
y = np.sin(x**2)

ax1.plot(x, y, 'c')
ax2.plot(x, y, 'b--')
```

```python
fig, axs = plt.subplots(2, 2)

x = np.linspace(0, 2 * np.pi)
y = np.sin(x**2)

axs[0, 0].set_title("Axis[0,0]")
axs[0, 0].plot(x, y, 'tab:orange')

axs[0, 1].set_title("Axis[0,1]")
axs[0, 1].plot(x, y, 'tab:green')

axs[1, 0].set_title("Axis[1,0]")
axs[1, 0].plot(x, y, 'tab:blue')

axs[1, 1].set_title("Axis[1,1]")
axs[1, 1].plot(x, y, 'tab:pink')

for ax in axs.flat:
    ax.set(xlabel = "x-axis", ylabel = "y-axis")

for ax in axs.flat:
    ax.label_outer()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (10, 5), dpi = 1000)

x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x**2)

for ax in axes:
    ax.plot(x, y, 'tab:pink')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title("tokyo")
    
fig = plt.tight_layout()    
fig.savefig("Matplotlib.png", dpi = 200)
    
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, color = "blue", alpha = 0.5,label = "x**2")
ax.plot(x, x**3, 'g--', label = "x**3")
ax.legend(loc = 2)
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize = (12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("Default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("Custom axes")

plt.scatter(x, y)
```
