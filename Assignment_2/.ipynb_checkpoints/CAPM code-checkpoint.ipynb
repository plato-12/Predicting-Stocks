```python
from scipy import stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import pandas_datareader.data as web
```

```python
start = "2021-01-01"
end = "2022-01-01"

df_n = web.DataReader(name='^NSEI', data_source='yahoo', start=start, end=end)
df_n.head()
```

```python
df_s = web.DataReader(name='TCS.NS', data_source='yahoo', start=start, end=end)
df_s.head()
# stock = tata consultancy services limited
```

```python
df_s['Close'].plot(label="Tata Consultancy Services Limited", figsize=(15,10))
df_n['Close'].plot(label="Nifty 50")
plt.legend()
plt.show()
```

```python
# wrt to start
df_s['c'] = df_s['Close']/df_s['Close'].iloc[0]
df_n['c'] = df_n['Close']/df_n['Close'].iloc[0]
```

```python
df_s['c'].plot(label="Tata Consultancy Services Limited", figsize=(15,10), color='pink', lw=4, ls='-')
df_n['c'].plot(label="Nifty 50", color='cyan', lw=4, ls='-')

plt.legend()
plt.show()
```

```python
df_s['change'] = df_s['Close'].pct_change(periods=1)
df_n['change'] = df_n['Close'].pct_change(periods=1)
```

```python
df_n.head()
```

```python
plt.style.use('dark_background')
plt.scatter(df_s['change'], df_n['change'])
```

```python
lin_reg = stats.linregress(df_s['change'].iloc[1:], df_n['change'].iloc[1:])
lin_reg
```

```python
beta, alpha, r_val, p_val, std_err = lin_reg
```
